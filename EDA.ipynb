FIRST OF ALL GONNA CHECK THE NUMBER OF FILES FOR BOTH CLASSES.

```python
import os

data_dir="kagglecatsanddogs_3367a/PetImages"
cats = len(os.listdir(os.path.join(data_dir,"Cat")))
dogs = len(os.listdir(os.path.join(data_dir,"Dog")))

print(f"Number of cat images :{cats}")
print(f"Number of dog images :{dogs}")
```

DATA IS ALMOST EQUALLY DISTRIBUTED. NOW CHECKING FOR CORROUPTED IMAGE.And CHECKING  THE SIZE OF THE IMAGES IMAGE.

```python
from PIL import Image

sizes = []

for cls in ["Cat", "Dog"]:
    folder = os.path.join(data_dir, cls)
    for file in os.listdir(folder):
        try:
            img = Image.open(os.path.join(folder, file))
            sizes.append(img.size)  # (width, height)
        except:
            pass  # skip corrupted

print(f"Sample sizes: {sizes[:5]}")
print(f"Min size: {min(sizes)}, Max size: {max(sizes)}")
```

CHECKING THE FILE IS CORROUPTED OR NOT.

```python
from PIL import Image

corroupted = []

for cls in ["Cat","Dog"]:
    folder=os.path.join(data_dir,cls)
    for file in os.listdir(folder):
        try:
            img = Image.open(os.path.join(folder,file))
            img.verify()
        except:
            corroupted.append(file)

print(f"corroupted file :{corroupted}")
```

NOW CHECKING THE ORIENTATION OF THE IMAGES LANDSCAPE SQUARE OR PORTRAIT

```python
from PIL import Image
import os

orientation_count = {
    "portrait": 0,
    "landscape": 0,
    "square": 0
}

ratios = []

for cls in ["Cat", "Dog"]:
    folder = os.path.join(data_dir, cls)
    for file in os.listdir(folder):
        try:
            with Image.open(os.path.join(folder, file)) as img:
                w, h = img.size
                ratio = w / h
                ratios.append(ratio)

                if w > h:
                    orientation_count["landscape"] += 1
                elif h > w:
                    orientation_count["portrait"] += 1
                else:
                    orientation_count["square"] += 1
        except:
            pass

print("Orientation distribution:", orientation_count)
print("Aspect ratio range:", min(ratios), "to", max(ratios))
```

CHECKING IS THERE ANY DUPLICATE VALUES OR NOT IN THE DATASET.

```python
import os
import hashlib
hashes={}
duplicated=[]
classes=["Cat","Dog"]
for cls in classes:
    folder = os.path.join(data_dir,cls)
    for file in os.listdir(folder):
        try:
            with open(file,"rb") as f:
                file_hash=hashlib.md5(f.read()).hexdigest()
                
            if file_hash in hashes:
                duplicated.append((file,hashes[file_hash]))
                
            else:
                hashes[file_hash]=file
                
        except:
            pass
```
